```python
def data_process(data, prev_data):
    
    prev_data=prev_data.rename(columns={'label_Q':'prev_label','X1_m3':'X1_pm3','X2_m3':'X2_pm3','X3_m3':'X3_pm3','X4_m3':'X4_pm3','X5_m3':'X5_pm3','X6_m3':'X6_pm3','X7_m3':'X7_pm3','X8_m3':'X8_pm3','B1_m3':'B1_pm3','B2_m3':'B2_pm3','B3_m3':'B3_pm3','B4_m3':'B4_pm3','B5_m3':'B5_pm3','B6_m3':'B6_pm3','B7_m3':'B7_pm3','X1_m1':'X1_pm1','X2_m1':'X2_pm1','X3_m1':'X3_pm1','X4_m1':'X4_pm1','X5_m1':'X5_pm1','X6_m1':'X6_pm1','X7_m1':'X7_pm1','X8_m1':'X8_pm1','X1_m2':'X1_pm2','X2_m2':'X2_pm2','X3_m2':'X3_pm2','X4_m2':'X4_pm2','X5_m2':'X5_pm2','X6_m2':'X6_pm2','X7_m2':'X7_pm2','X8_m2':'X8_pm2','E1_Q':'E1_pQ','E2_Q':'E2_pQ','E3_Q':'E3_pQ','E4_Q':'E4_pQ','E5_Q':'E5_pQ','E6_Q':'E6_pQ','E7_Q':'E7_pQ','E8_Q':'E8_pQ','E9_Q':'E9_pQ','E10_Q':'E10_pQ','E11_Q':'E11_pQ','E12_Q':'E12_pQ','E13_Q':'E13_pQ','E14_Q':'E14_pQ','E16_Q':'E16_pQ','E18_Q':'E18_pQ'})
    data=pd.merge(data,prev_data[['cust_no','prev_label','X1_pm3','X2_pm3','X3_pm3','X4_pm3','X5_pm3','X6_pm3','X7_pm3','X8_pm3','X1_pm2','X2_pm2','X3_pm2','X4_pm2','X5_pm2','X6_pm2','X7_pm2','X8_pm2','X1_pm1','X2_pm1','X3_pm1','X4_pm1','X5_pm1','X6_pm1','X7_pm1','X8_pm1','B1_pm3','B2_pm3','B3_pm3','B4_pm3','B5_pm3','B6_pm3','B7_pm3','E1_pQ','E2_pQ','E3_pQ','E4_pQ','E5_pQ','E6_pQ','E7_pQ','E8_pQ','E9_pQ','E10_pQ','E11_pQ','E12_pQ','E13_pQ','E14_pQ','E16_pQ','E18_pQ']],on='cust_no', how='left', validate='one_to_one')
    #各月末存款总额
    data['sum_X_m1']=data.X1_m1+data.X2_m1+data.X3_m1+data.X4_m1+data.X5_m1+data.X6_m1+data.X8_m1
    data.sum_X_m1[data.sum_X_m1==0]=1

    data['sum_X_m2']=data.X1_m2+data.X2_m2+data.X3_m2+data.X4_m2+data.X5_m2+data.X6_m2+data.X8_m2
    data.sum_X_m2[data.sum_X_m2==0]=1

    data['sum_X_m3']=data.X1_m3+data.X2_m3+data.X3_m3+data.X4_m3+data.X5_m3+data.X6_m3+data.X8_m3
    data.sum_X_m3[data.sum_X_m3==0]=1



    #各类存款占余额比例
    for j in range(1, 4):
        for i in range(1, 9):
            data['X'+str(i)+'pct_m'+str(j)]=data['X'+str(i)+'_m'+str(j)]/data['sum_X_m'+str(j)]
    #         data['X'+str(i)+'pct_m'+str(j)].fillna(


    for i in range(1, 9):
#         print(i)
        data['X'+str(i)+'_D_m3']=data['X'+str(i)+'_m3']-data['X'+str(i)+'_pm3']
        data['X'+str(i)+'_Dpct_m3']=data['X'+str(i)+'_D_m3']/(data['X'+str(i)+'_pm3']+data['X'+str(i)+'_m3'])
    #         data['X'+str(i)+'pct_m'+str(j)].fillna(

    for i in range(1, 9):
        data['X'+str(i)+'_D2_m3']=data['X'+str(i)+'_m3']-data['X'+str(i)+'_pm1']
        data['X'+str(i)+'_D2pct_m3']=data['X'+str(i)+'_D2_m3']/(data['X'+str(i)+'_pm1']+data['X'+str(i)+'_m3'])
    #         data['X'+str(i)+'pct_m'+str(j)].fillna(

    for i in range(1, 9):
        data['X'+str(i)+'_D3_m3']=data['X'+str(i)+'_m3']-data['X'+str(i)+'_m1']
        data['X'+str(i)+'_D3pct_m3']=data['X'+str(i)+'_D3_m3']/(data['X'+str(i)+'_m1']+data['X'+str(i)+'_m3'])
    #         data['X'+str(i)+'pct_m'+str(j)].fillna(


    for i in range(1, 6):
        data['B'+str(i)+'_D_m3']=data['B'+str(i)+'_m3']-data['B'+str(i)+'_pm3']
        data['B'+str(i)+'_Dpct_m3']=data['B'+str(i)+'_D_m3']/(data['B'+str(i)+'_pm3']+data['B'+str(i)+'_m3'])
        data['B'+str(i)+'_D2_m3']=data['B'+str(i)+'_m3']-data['B'+str(i)+'_m1']
        data['B'+str(i)+'_D2pct_m3']=data['B'+str(i)+'_D2_m3']/(data['B'+str(i)+'_m3']+data['B'+str(i)+'_m1'])
        

    i=7
    data['B'+str(i)+'_D_m3']=data['B'+str(i)+'_m3']-data['B'+str(i)+'_pm3']
    data['B'+str(i)+'_Dpct_m3']=data['B'+str(i)+'_D_m3']/(data['B'+str(i)+'_pm3']+data['B'+str(i)+'_m3'])
#     data['B'+str(i)+'_D2_m3']=data['B'+str(i)+'_m3']-data['B'+str(i)+'_m1']
#     data['B'+str(i)+'_D2pct_m3']=data['B'+str(i)+'_D2_m3']/(data['B'+str(i)+'_m3']+data['B'+str(i)+'_m1'])
    data['B6_m3']=data['B6_m3'].astype('datetime64[ns]')
    data['B6_pm3']=data['B6_pm3'].astype('datetime64[ns]')
    
    data['B6_delt_Q']=(datetime.today()-data['B6_m3']).apply(lambda x: x.days)
    data['B6_D_Q']=(data['B6_m3']-data['B6_pm3']).apply(lambda x: x.days)
    
    for i in range(1,15):
        data['E'+str(i)+'_Q']=data['E'+str(i)+'_Q'].astype('datetime64[ns]')
        data['E'+str(i)+'_pQ']=data['E'+str(i)+'_pQ'].astype('datetime64[ns]')
        data['E'+str(i)+'_D_Q']=(data['E'+str(i)+'_Q']-data['E'+str(i)+'_pQ']).apply(lambda x: x.days)
    i=16
    data['E'+str(i)+'_Q']=data['E'+str(i)+'_Q'].astype('datetime64[ns]')
    data['E'+str(i)+'_pQ']=data['E'+str(i)+'_pQ'].astype('datetime64[ns]')
    data['E'+str(i)+'_D_Q']=(data['E'+str(i)+'_Q']-data['E'+str(i)+'_pQ']).apply(lambda x: x.days)
    i=18
    data['E'+str(i)+'_Q']=data['E'+str(i)+'_Q'].astype('datetime64[ns]')
    data['E'+str(i)+'_pQ']=data['E'+str(i)+'_pQ'].astype('datetime64[ns]')
    data['E'+str(i)+'_D_Q']=(data['E'+str(i)+'_Q']-data['E'+str(i)+'_pQ']).apply(lambda x: x.days)
    
    #3个月各类存款总额，也可以看成各类存款3个月末的均值（只要除以3）
    for i in range(1,9):
        data['X'+str(i)+'_3mt']=data['X'+str(i)+'_m1']+data['X'+str(i)+'_m2']+data['X'+str(i)+'_m3']
        data['X'+str(i)+'_p3mt']=data['X'+str(i)+'_pm1']+data['X'+str(i)+'_pm2']+data['X'+str(i)+'_pm3']
        data['X'+str(i)+'_D_3mt']=data['X'+str(i)+'_3mt']-data['X'+str(i)+'_p3mt']
        data['X'+str(i)+'_Dpct_3mt']=data['X'+str(i)+'_D_3mt']/(data['X'+str(i)+'_p3mt']+data['X'+str(i)+'_3mt'])

    #所有存款3个月总额，或者看成所有存款3个月末额均值（除以3）。如果为0则变成1（后面要做分母）
    data['sum_X_3mt']=data.X1_3mt+data.X2_3mt+data.X3_3mt+data.X4_3mt+data.X5_3mt+data.X6_3mt+data.X8_3mt
    data.sum_X_3mt[data.sum_X_3mt==0]=1


    #各类存款3个月总额占存款总额的比例
    for i in range(1, 9):
        data['X'+str(i)+'pct_3mt']=data['X'+str(i)+'_3mt']/data['sum_X_3mt']



    #128是各类定期、结构、大额存单类总和（3个月加总），456是各类理财、基金、资管投资类总和（3个月加总）。各类占存款总额的比例。
    data['X128pct_3mt']=(data.X1_3mt+data.X2_3mt+data.X8_3mt)/data.sum_X_3mt
    data['X456pct_3mt']=(data.X4_3mt+data.X5_3mt+data.X6_3mt)/data.sum_X_3mt

    #三个月内转出转入总笔数和总金额
    # data['B2_m3']=data.B2_m1+data.B2_m2+data.B2_m3
    # data['B3_m3']=data.B3_m1+data.B3_m2+data.B3_m3
    # data['B4_m3']=data.B4_m1+data.B4_m2+data.B4_m3
    # data['B5_m3']=data.B5_m1+data.B5_m2+data.B5_m3

    #转入转出次数金额比例
    data['B2_pct_m3']=data['B2_m3']/(data['B2_m3']+data['B4_m3'])
    data['B4_pct_m3']=data['B4_m3']/(data['B2_m3']+data['B4_m3'])

    data['B3_pct_m3']=data['B3_m3']/(data['B3_m3']+data['B5_m3'])
    data['B5_pct_m3']=data['B5_m3']/(data['B3_m3']+data['B5_m3'])

    #转入转出占余额比例
    data['B3_X_pct_m3']=data['B3_m3']/data['sum_X_m3']
    data['B5_X_pct_m3']=data['B5_m3']/data['sum_X_m3']

    #平均单笔转出转入金额
    data['B3_B2_m3']=data['B3_m3']/data['B2_m3']
    data['B5_B4_m3']=data['B5_m3']/data['B4_m3']

    #平均单笔占余额比例
    data['B3_B2_X_m3']=data['B3_B2_m3']/data['sum_X_m3']
    data['B5_B4_X_m3']=data['B5_B4_m3']/data['sum_X_m3']

    #净转入
    data['Net_B2_m3']=data['B2_m3']-data['B4_m3']
    data['Net_B3_m3']=data['B3_m3']-data['B5_m3']

    #净转入占比
    data['Net_B2_pct_m3']=data['Net_B2_m3']/data['B7_m3']
    data['Net_B3_pct_m3']=data['Net_B2_m3']/data['sum_X_m3']


    #把E1到E14全部转为到今天的日期差
    for i in range(1,15):
    #     print(i)
        data['E'+str(i)+'_Q']=data['E'+str(i)+'_Q'].astype('datetime64[ns]')
        data['E'+str(i)+'_delt_Q']=(datetime.today()-data['E'+str(i)+'_Q']).apply(lambda x: x.days)

    data['E16_Q']=data['E16_Q'].astype('datetime64[ns]')
    data['E16_delt_Q']=(datetime.today()-data['E16_Q']).apply(lambda x: x.days)
    data['E18_Q']=data['E18_Q'].astype('datetime64[ns]')
    data['E18_delt_Q']=(datetime.today()-data['E18_Q']).apply(lambda x: x.days)

    #缺失值比较多的几个E转成标签，有日期就0，没日期就1
    for i in range(4, 15):
        data.loc[data['E'+str(i)+'_delt_Q'].isna(), 'E'+str(i)+'_lab']=1
        data.loc[-data['E'+str(i)+'_delt_Q'].isna(), 'E'+str(i)+'_lab']=0
        data.loc[data['E'+str(i)+'_delt_Q'].isna(), 'E'+str(i)+'_delt_Q']=data['E'+str(i)+'_delt_Q'].mean()


    data.loc[data['E16_delt_Q'].isna(), 'E16_lab']=1
    data.loc[-data['E16_delt_Q'].isna(), 'E16_lab']=0
    data.loc[data['E16_delt_Q'].isna(), 'E16_delt_Q']=data['E16_delt_Q'].mean()

    data.loc[data['E18_delt_Q'].isna(), 'E18_lab']=1
    data.loc[-data['E18_delt_Q'].isna(), 'E18_lab']=0
    data.loc[data['E18_delt_Q'].isna(), 'E18_delt_Q']=data['E18_delt_Q'].mean()

    data['I1_Q']=data.I1_Q.fillna('未知')
    data['I5_Q']=data.I5_Q.fillna('未知')
    data['I10_Q']=data.I10_Q.fillna('未知')
    data['I13_Q']=data.I13_Q.fillna('未说明的婚姻状况')
    data['I14_Q']=data.I14_Q.fillna('未知')

    data['C1_3mt']=data['C1_m1']+data['C1_m2']+data['C1_m3']
    data['C2_3mt']=data['C2_m1']+data['C2_m2']+data['C2_m3']

    for i in range(2,15):
        data['E'+str(i)+'D_Q']=data['E'+str(i)+'_delt_Q']-data['E1_delt_Q']
        
    i=16
    data['E'+str(i)+'D_Q']=data['E'+str(i)+'_delt_Q']-data['E1_delt_Q']
    i=18
    data['E'+str(i)+'D_Q']=data['E'+str(i)+'_delt_Q']-data['E1_delt_Q']
    
#     data['C_Q']=data['E'+str(i)+'_delt_Q']-data['E1_delt_Q']
    
    data.prev_label.fillna(1, inplace=True)

    
    return data
    # features=['X1_m3','X2_m3','X3_m3','X4_m3','X5_m3','X6_m3','X7_m3','X8_m3','X1_D_m3','X2_D_m3','X3_D_m3','X4_D_m3','X5_D_m3','X6_D_m3','X7_D_m3','X8_D_m3','E15_Q','E17_Q','I1_Q','I2_Q','I3_Q','I4_Q','I5_Q','I6_Q','I7_Q','I8_Q','I10_Q','I12_Q','I13_Q','I14_Q','I15_Q','I16_Q','I17_Q','I18_Q','I19_Q','I20_Q','E1_delt_Q','E2_delt_Q','E3_delt_Q','E4_delt_Q','E5_delt_Q','E6_delt_Q','E7_delt_Q','E8_delt_Q','E9_delt_Q','E10_delt_Q','E12_delt_Q','E13_delt_Q','E14_delt_Q','E16_delt_Q','E18_delt_Q','sum_X_m1','sum_X_m2','sum_X_m3','X1pct_m3','X2pct_m3','X3pct_m3','X4pct_m3','X5pct_m3','X6pct_m3','X7pct_m3','X8pct_m3','X1_3mt','X2_3mt','X3_3mt','X4_3mt','X5_3mt','X6_3mt','X7_3mt','X8_3mt','sum_X_3mt','X1pct_3mt','X2pct_3mt','X3pct_3mt','X4pct_3mt','X5pct_3mt','X6pct_3mt','X7pct_3mt','X8pct_3mt','X128pct_3mt','X456pct_3mt','B2_m3','B3_m3','B4_m3','B5_m3','B2_pct_m3','B4_pct_m3','B3_pct_m3','B5_pct_m3','B3_X_pct_m3','B5_X_pct_m3','B3_B2_m3','B5_B4_m3','B3_B2_X_m3','B5_B4_X_m3','Net_B2_m3','Net_B3_m3','Net_B2_pct_m3','Net_B3_pct_m3','C1_3mt','C2_3mt','prev_label']
    # for i in features:
    #     if len(np.where(data[i].isna())[0])>0:
    #         data.loc[data[i].isna(), i]=data[i].mean()
    #     print(i, len(np.where(data[i].isna())[0]))
```
